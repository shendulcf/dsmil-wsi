1. WSI 目录设置

```python
import os 
import sys
import shutil
from tqdm import tqdm

# svs_path = r'/home/sci/Disk2/tcga_brca/WSI'
svs_path = r'/home/sci/Disk_data/tcga_nsclc_sub/WSI'
patch_path = r'/home/sci/Disk_data/tcga_nsclc_sub/WSI/single/sci'
slide_list = tqdm(os.listdir(svs_path))
# print(slide_list)
if not os.path.exists(os.path.join(svs_path,'tcga_nsclc','luad')):
    normal_path = os.path.join(svs_path, 'tcga_nsclc','luad')
    os.makedirs(normal_path)
if not os.path.exists(os.path.join(svs_path,'tcga_nsclc','lusc')):
    tumor_path = os.path.join(svs_path, 'tcga_nsclc','lusc')
    os.makedirs(tumor_path)

for slide in slide_list:
    slide_list.set_description('Precessing %s' % slide)
    file_path = os.path.join(svs_path, slide)
    if slide[13] == '0':
        shutil.move(file_path,tumor_path)
    else:
        shutil.move(file_path,normal_path)
        
```
